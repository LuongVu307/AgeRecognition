```python
from tensorflow.keras.models import load_model
import cv2
import os
from os.path import join
import numpy as np
import matplotlib.pyplot as plt

prediction_dct = {0 : '25-30', 1 : '42-48',2 : '6-20',3 : '60-98'}

MAIN_PATH = os.getcwd()
model = load_model(join(MAIN_PATH, "AgeRecognition.h5"))


image = cv2.imread(join(MAIN_PATH, 'exmaple (1).png'))
image = cv2.resize(image, (200, 200))

final_image = np.expand_dims(image, 0)
prediction = model.predict(final_image)

print("FINAL_PREDICTION: YOU'RE ", prediction_dct[np.argmax(prediction)], "YEARS OLD")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
